```python
#https://www.youtube.com/watch?v=QhD015WUMxE

#IMPORT LIBRARIES
import requests
from bs4 import BeautifulSoup

#REQUEST WEBPAGE AND STORE IT AS A VARIABLE
page_to_scrape = requests.get("http://quotes.toscrape.com")

#USE BEAUTIFULSOUP TO PARSE THE HTML AND STORE IT AS A VARIABLE
soup = BeautifulSoup(page_to_scrape.text, 'html.parser')

#FIND ALL THE ITEMS IN THE PAGE WITH A CLASS ATTRIBUTE OF 'TEXT'
#AND STORE THE LIST AS A VARIABLE
quotes = soup.findAll('span', attrs={'class':'text'})

#FIND ALL THE ITEMS IN THE PAGE WITH A CLASS ATTRIBUTE OF 'AUTHOR'
#AND STORE THE LIST AS A VARIABLE
authors = soup.findAll('small', attrs={"class":"author"})

#LOOP THROUGH BOTH LISTS USING THE 'ZIP' FUNCTION
#AND PRINT AND FORMAT THE RESULTS
for quote, author in zip(quotes, authors):
    print(quote.text + "-" + author.text)
```

```python
# this won't work since the website isn't using static HTML...it's pulling in elements dynamically using javascript
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the page to scrape
url = "https://sam.gov/search/?page=1&pageSize=100&sort=&index=sca&sfm%5Bstatus%5D%5Bis_active%5D=true&sfm%5BwdLocationWrapper%5D%5BwdStates%5D%5B0%5D%5Bkey%5D=MD&sfm%5BwdLocationWrapper%5D%5BwdStates%5D%5B0%5D%5Bvalue%5D=District%20of%20Columbia"

# Send a request to the webpage
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.content, 'html.parser')

    # Initialize lists to store the data
    states = []
    counties = []
    wd_numbers = []

    # Find the relevant elements that contain the data
    wd_items = soup.find_all('div', class_='ng-star-inserted')

    for item in wd_items:
        # Extract the state, county, and WD #
        state = item.find('span', class_='sds-field__value').text.strip() if item.find('span', class_='sds-field__value') else None
        county = item.find('span', class_='sds-field__value').text.strip() if item.find('span', class_='sds-field__value') else None
        wd_number = item.find('span', class_='usa-link').text.strip() if item.find('span', class_='usa-link') else None

        if state and county and wd_number:
            states.append(state)
            counties.append(county)
            wd_numbers.append(wd_number)

    # Create a DataFrame
    data = {'State': states, 'County': counties, 'Service Contract Act WD #': wd_numbers}
    df = pd.DataFrame(data)

    # Display the DataFrame
    print(df)
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")

```

```python
##the "final" split, which looks at the chunks from the bulk page and pulls out state as well as county

dfNewEngland = pd.DataFrame(columns=['search','wd num','state', 'county', 'date', 'revision_num'])

wd_elements = soup.find_all('div', class_ = {"sds-field__value","sds-field__name", "grid-col"})
wd_list = [tag.text for tag in wd_elements]
data = '\n'.join(wd_list)

# Regular expression pattern to match each chunk
pattern = r"(Service Contract Act WD #:.*?Published Date\n.*?\d{4})"

# Find all chunks using the pattern
chunks = re.findall(pattern, data, re.DOTALL)

# Initialize list to store data
rows = []

# Process each chunk to extract information
for chunk in chunks:

    # Extract WD number
    wd_number_match = re.search(r"Service Contract Act WD #:\s*(\S+)", chunk)
    wd_number = wd_number_match.group(1) if wd_number_match else None

    # Extract published date
    published_date_match = re.search(r"Published Date\n(.*)", chunk)
    published_date = published_date_match.group(1).strip() if published_date_match else None

    # Extract revision number
    revision_number_match = re.search(r"Revision Number\n(\d+)", chunk)
    revision_number = revision_number_match.group(1) if revision_number_match else None

    # Split text using the delimiters ":" and "," and "\n" and the word "in" and the sudden capitalized letters in the middle of location_description
    location_description = re.split(r'[:,\n-]|(?<=[a-z])(?=[A-Z]| in )', chunk)

    # Remove any leading/trailing whitespace from each item
    location_description = [word.strip() for word in location_description if word.strip()]

    new_england = ['Connecticut', 'Maine','Massachusetts', 'New Hampshire', 'Rhode Island', 'Vermont']
    state_pattern = re.compile(r'\b(' + '|'.join(new_england) + r')\b', re.IGNORECASE)
    ne_rows = []

    for i in range(len(location_description)):
        line = location_description[i].strip()

        if state_pattern.search(line):
            state = state_pattern.search(line).group(0)
        elif re.search(r'county', line, re.IGNORECASE):
            county = location_description[i]
            county = line.replace("in ", "").replace(".", "")
            county = re.sub(r'\bcounty\b', '', county, flags=re.IGNORECASE).strip()
            ne_rows.append({'state':state, 'county': county, 'wd num':wd_number, 'date':published_date, 'revision_num':revision_number})
#need to make date into obj
#need to clean up names
#some WDs aren't being captured (2015-2049,2015-2053)...maybe the append should trigger per WD
# too many rows (there's an extra one per county...should be 1 extra row per STATE)
# how to capture "New Haven county (or counties that are 2 words long? Maybe replace the normal delimiter with a "SUPER DELIMITER" and then use that super delimiter in the search)"
    dfAppend3 = pd.DataFrame(ne_rows)
    dfNewEngland = pd.concat([dfNewEngland, dfAppend3], ignore_index=True)
    dfNewEngland = dfNewEngland.groupby(['wd num', 'state', 'revision_num'], as_index=False).agg({
    'county': ', '.join,
    'search': 'first',
    'date': 'first'
})

print(dfNewEngland)
```
